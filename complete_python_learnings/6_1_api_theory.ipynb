### A REST API (Representational State Transfer) is a way for two computer systems to communicate using the HTTP technologies found in web browsers and servers. Sharing data between two or more systems has always been a fundamental requirement of software development.

```python
# code of REST API 

from flask import Flask , request , jsonify
app = Flask(__name__)

@app.route("/function")
def function1():
    test1 = request.args.get('val1')
    test2 = request.args.get('val2')
    return '''<h1> My result is : {} , {}</h1>'''.format(test1,test2)
if __name__ == '__main__':
    app.run()
```
