```python
import os

from flask import Flask, render_template, request, jsonify
from flask_cors import CORS, cross_origin # this is for deployement purpose(just mentioned)
import requests # requesting url/data
from bs4 import BeautifulSoup as bs # for beautification
from urllib.request import urlopen as uReq #open up url
```

```python
# stroing url to flipkart variable.
flipkart_url = "https://www.flipkart.com/search?q=" + 'iphone13'
flipkart_url
```

```python
# requests url(urlib.requwsts)
uClient = uReq(flipkart_url)
uClient
```

```python
# It will get the data related to your flipkart url html
flipkartPage = uClient.read()
flipkartPage
```

```python
uClient.close()
```

```python
# converts into strucutured format not well structured.
flipkart_html = bs(flipkartPage, "html.parser")
flipkart_html
```

```python
# get the information of all product:
all_products = flipkart_html.findAll("div", {"class" : "_1AtVbE col-12-12"})
all_products
```

```python
# length of products(total products) :
len(all_products)
```

```python
del all_products [0:3]
```

```python
# length of products(total products) :
len(all_products)
```

```python
# first product(first product details) :
product1 = all_products[0]
product1
```

```python
# getting product link :
product_link = "https://www.flipkart.com" + product1.a['href']
product_link
```

```python
# response, whether it may be 200, 404, 500
response = requests.get(product_link)
response.encoding = 'utf-8'
response
```

```python
# converts all product details into text format
product_html = bs(response.text, 'html.parser')
print(product_html)
```

## get inside the link and to get the reviews of the products :

```python
# from product_html(text format) getting all comments
comments = product_html.findAll("div", {'class' : '_16PBlm'})
comments
```
