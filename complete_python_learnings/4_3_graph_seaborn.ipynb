# seaborn plots :

### Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.
### Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures. Seaborn helps you explore and understand your data.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# relation between tip and total_bill
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips)
```

```python
# tip and total bill received from smokers and non-smokers
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips, hue = 'smoker')
```

```python
# days recieved tip from sex
sns.relplot(x = 'tip' , y = 'day' , data = tips, hue = 'sex')
```

```python
# check smokers and non-smokers:
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips, hue = 'day', style = 'smoker')
```

```python
# based on size, seperate smokers and non-smokers, who paid bill and tip
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips, hue = 'day', style = 'smoker', size = 'size')
```

```python
# random dataset
dff = pd.DataFrame(np.random.randn(400, 3), columns = (['A', 'B', 'C']))
dff
```

```python
sns.relplot(x = "A", y = "B", kind = 'line', data = dff, sort = True)
```

```python
sns.relplot(x = 'A', y = 'B', data = dff , hue = 'C')
```

## distplot : used to check disributed data

```python
# random data:
x = np.random.randn(500)
x
```

```python
# is this normally distributed or not:
sns.distplot(x)
```

```python
# If we increase size of random numbers, it is more normally distributed.
x1 = np.random.randn(1000)
sns.distplot(x1)
```

```python
# If we increase size of random numbers, it is more normally distributed.
x2 = np.random.randn(10000)
sns.distplot(x2)
```

#### As we have maximum number of data, we'll get good normally distributed dataset

## jointplot : combination of two plots

```python
# tips dataset
sns.jointplot(x = tips.total_bill , y = tips.tip)
```

### Right skewed histogram :
#### A histogram in which most of the data falls to the right of the graph's peak is known as a right-skewed histogram. It is also known as a positively skewed histogram.  Medain > Mean
### left skewed histogram :
#### A distribution is called skewed left if, as in the histogram above, the left tail (smaller values) is much longer than the right tail (larger values). Note that in a skewed left distribution, the bulk of the observations are medium/large, with a few observations that are much smaller than the rest.  Mean > Medain
