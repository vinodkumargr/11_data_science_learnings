## Permforming EDA on FitBit dataset to predict the calories 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas_profiling
```

```python
# imporrting dataset
df = pd.read_csv('FitBit data.csv')
df
```

```python
#nullvalues.sum
df.isnull().sum()
```

```python
# get the Id's of user whose TotalStep count < 50
df.Id[df.TotalSteps < 50]
```

```python
df.dtypes
```

```python
# unique number of dates :
df.ActivityDate.unique()
```

```python
# seperate date, month and year :
df['date'] = pd.DatetimeIndex(df.ActivityDate)
df
```

```python
df.dtypes
```

```python
# create day column
df['date'] = pd.DatetimeIndex(df.ActivityDate).day
df
```

```python
# create month column
df['month'] = pd.DatetimeIndex(df.ActivityDate).month
df
```

```python
# create year column
df['year'] = pd.DatetimeIndex(df.ActivityDate).year
df
```

```python
# getting 4 columns:
df[['ActivityDate','date', 'month', 'year']]
```

```python
# boxplot of date and calorie :
plt.figure(figsize=(18,9))
sns.boxplot(x='date', y='Calories', data=df)
```

```python
# created week column
df['week'] = pd.DatetimeIndex(df.ActivityDate).week
df
```

```python
# to create day number
# df['day_name'] = pd.to_datetime(df.ActivityDate).dt.weekday
```

```python
# create a day name column
df['day_name'] = pd.to_datetime(df.ActivityDate).dt.day_name()
df
```

```python
# boxplot of day_name and calorie :
plt.figure(figsize=(18,9))
sns.boxplot(x='day_name', y='Calories', data=df)
```

```python
plt.figure(figsize=(19,9))
sns.scatterplot(x='Calories', y='TotalSteps', data=df)
```
