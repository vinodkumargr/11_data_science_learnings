```python
df = pd.DataFrame( {
    'col1' : [1,2,3,4,5],
    'col2' : [45,56,67,34,34],
    'col3' : 'vinod kumar sai abhi shreyas'.split()  # this split helps in seperate names after a spaceas unique names.
} )

df
```

```python
# create a new column, col4 and values should be square of col1

df['col4'] = df.col1**2
df
```

```python
# or you can create a new function 

def test(a):
    return a**2


df['col4_4'] = df.col1.apply(test)

df  # this is also retens same as pd.col1**2

# apply will take function and returns output
```

```python
# create col5, values should be first character/word of col3

df['col5'] = df.col3.str[0]
df
```

```python
# create col6, if col5 == s, then value shld be s, if not value shld be z.

def test1(a):
    if a == 's':
        return 's'
    else:
        return 'z'

df['col6'] = df.col5.apply(test1)
df
```

```python
# you can use lambda function here : 

df['col6_6'] = df.col5.apply(lambda a: 's' if a == 's' else 'z')
df
```

```python
# create col7, where this is length of a col3

df['col7'] = df.col3.apply(lambda a : len(a))
df
```

```python
# creaye col8, values shld be log(col2) :

import math  # in math library in-built log function is there

df['col8'] = df.col2.apply(lambda a : math.log(a))
df
```
