```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

## date_range :
#### date_range() is one of the general functions in Pandas which is used to return a fixed frequency DatetimeIndex. Syntax: pandas.date_range(start=None, end=None, periods=None, freq=None, tz=None, normalize=False, name=None, closed=None, **kwargs)

```python
df = pd.DataFrame(np.random.randn(1000), columns = ['data'], index = pd.date_range('19/9/2022', periods = 1000))
df
```

```python
df.plot(figsize = (15,5)) # changes figure size
```

```python
## dataset from github
dff = pd.read_csv("https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/master/Iris.csv")
dff
```

```python
dff.plot(figsize = (16,6))# i can't see the graph clearly, i'll rop id column
```

```python
# i"m going to drop id column :
dff = dff.drop ('Id', axis = 1, )
dff
```

## line graph : default graph

```python
dff.plot(figsize = (16,6))
```

## bar graph :

```python
dff.plot(figsize = (16,6), kind = 'bar')
```

```python
# data for below bar plot.
dff.iloc[5:11]
```

```python
# get bar plot from 5 to 10th row :
dff.iloc[5:11].plot(figsize = (16,6), kind = 'bar')
```

## barh -> bar-horizantal :

```python
dff.iloc[5:11].plot(figsize = (16,6), kind = 'barh')
```

## hist -> histogram graph :
#### A histogram is a display of statistical information that uses rectangles to show the frequency of data items in successive numerical intervals of equal size.

```python
dff.plot(kind = "hist", figsize = (12,4))
```

```python
# only one column :
dff.SepalLengthCm.plot(kind ='hist')
```

```python
# to chnage the x to y and y to x a xis :
dff.SepalLengthCm.plot(kind ='hist', orientation = 'horizontal')
```

```python
# we are describing th e columns with color :
dff.hist(figsize = (15,10), color = 'black', alpha = .9) # alpha is contrast 
```

## box plot :
#### A box plot is a graphical rendition of statistical data based on the minimum, first quartile, median, third quartile, and maximum.

```python
dff.describe()
```

```python
dff.plot(kind = 'box', figsize = (15,6),color = {'boxes' : 'black','whiskers' : 'red'})
```

```python
dff.plot(kind = 'box', figsize = (15,6),color = {'boxes' : 'black','whiskers' : 'red'}, vert = False)
# vert = False, no vertical
```

## area graph :

```python
dff.plot(kind = 'area', figsize = (15,6))
```

```python
dff.plot(kind = 'area', figsize = (15,6), stacked = False) # stacked : Don't overloap one on another.
```

## scatter plot : (dff.plot.scatter)
#### a type of data visualization technique that shows the relationship between two numerical variables.

```python
dff
```

```python
dff.plot.scatter(x = 'SepalLengthCm' , y = 'PetalLengthCm', figsize = (17,6))
```

```python
dff.plot.scatter(x = 'SepalLengthCm' , y = 'PetalWidthCm', figsize = (17,6))
```

```python
dff.plot.scatter(x = 'SepalLengthCm' , y = 'PetalLengthCm' , c = 'SepalWidthCm', figsize = (15, 5))
# c is a color column
```

```python
dff.plot.scatter(x = 'SepalLengthCm' , y = 'PetalLengthCm' , c = 'SepalWidthCm', figsize = (20, 5), s = 400)
# s is size of dots :
```

```python
dff.plot.scatter(x = 'SepalLengthCm' , y = 'PetalLengthCm' , c = 'SepalWidthCm', figsize = (20, 5), s = dff['PetalWidthCm']*200)
# size depends on the petalwidth(according to petalwidth size varies)
```

## Hexbin graph :
#### represent the relationship of 2 numerical variables when you have a lot of data points. Without overlapping of the points, the plotting window is split into several hexbins. The color of each hexbin denotes the number of points in it.

```python
dff.plot.hexbin(x = 'SepalLengthCm' , y = 'PetalWidthCm', C = 'SepalWidthCm' , gridsize = 18, figsize = (10,6))
# C is color of column, gridsize issize of that data
```
