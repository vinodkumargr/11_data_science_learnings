```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('Data_Train.xlsx')
df.head(4)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# spliting date_of _journey to date, month and year columns
df['journey_date']=df['Date_of_Journey'].str.split('/').str[0]
df.head(4)
```

```python
df['journey_month']=df['Date_of_Journey'].str.split('/').str[1]
df.head(4)
```

```python
df['journey_year']=df['Date_of_Journey'].str.split('/').str[2]
df.head(4)
```

```python
# droping date_of_journey 
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head(2)
```

```python
# converting date, month and year columns into int
df['journey_date'] = df['journey_date'].astype(int)
df.info()
```

```python
df['journey_month'] = df['journey_month'].astype(int)
df['journey_year'] = df['journey_year'].astype(int)
df.info()
```

```python
df.head(2)
```

```python
# spliting dep_time into hour and minute
df['dep_hour']=df['Dep_Time'].str.split(':').str[0]
df.head(2)
```

```python
df['dep_minute']=df['Dep_Time'].str.split(':').str[1]
df.head(2)
```

```python
# after, drop dep_time column
df.drop('Dep_Time', axis =1, inplace=True)
df.head(2)
```

```python
# there is no info in additional_features column so drop it
df.drop('Additional_Info', axis=1, inplace=True)
df.head(2)
```

```python
# getting number of stops
df['total_stops']=df['Total_Stops'].str.split(' ').str[0]
df.head(2)
```

```python
# deleting original Total_Stops
df.drop('Total_Stops', axis=1, inplace=True)
df.head(2)
```
